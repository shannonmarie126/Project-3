```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
rent_data = Path("resources/rents.csv")
mortgage_data=Path("resources/mortgages.csv")
```

```python
rent_df= pd.read_csv(rent_data)
mortgage_df=pd.read_csv(mortgage_data)
rent_df.head()
```

```python
columns_to_keep=["RegionID", "RegionName", "StateName","2020-05-31","2021-05-31","2022-05-31","2023-05-31","2024-05-31"]
rent_clean_df = rent_df[columns_to_keep]
mortgage_clean_df = mortgage_df[columns_to_keep]
```

```python
rent_clean_df.drop(0, inplace=True)
rent_clean_df.head()
```

```python
mortgage_clean_df.drop(0,inplace=True)
mortgage_clean_df.head()
```

```python
rent_clean_df.replace("",np.nan, inplace=True)
rent_clean_df.dropna(inplace=True)
rent_clean_df.tail()
```

```python
mortgage_clean_df.replace("",np.nan, inplace=True)
mortgage_clean_df.dropna(inplace=True)
mortgage_clean_df.tail()
```

```python
rent_clean_df=rent_clean_df.rename(columns= {
    "RegionID":"region_id",
    "RegionName":"region_name",
    "StateName":"state_name",
    "2020-05-31":"2020",
    "2021-05-31":"2021",
    "2022-05-31":"2022",
    "2023-05-31":"2023",
    "2024-05-31":"2024"
})
rent_clean_df.head()
```

```python
mortgage_clean_df=mortgage_clean_df.rename(columns= {
    "RegionID":"region_id",
    "RegionName":"region_name",
    "StateName":"state_name",
    "2020-05-31":"2020",
    "2021-05-31":"2021",
    "2022-05-31":"2022",
    "2023-05-31":"2023",
    "2024-05-31":"2024"
})
mortgage_clean_df.head()
```

```python
rent_clean_df.to_csv("output/clean_rents.csv", index=False)
mortgage_clean_df.to_csv("output/clean_mortgages.csv",index=False)
```
