```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
rent_data = Path("resources/rents.csv")
mortgage_data=Path("resources/mortgages.csv")
```

```python
rent_df= pd.read_csv(rent_data)
mortgage_df=pd.read_csv(mortgage_data)
rent_df.head()
```

```python
columns_to_average_2020=["2020-01-31","2020-02-29","2020-03-31","2020-04-30","2020-05-31","2020-06-30","2020-07-31","2020-08-31","2020-09-30","2020-10-31","2020-11-30","2020-12-31"]
columns_to_average_2021=["2021-01-31","2021-02-28","2021-03-31","2021-04-30","2021-05-31","2021-06-30","2021-07-31","2021-08-31","2021-09-30","2021-10-31","2021-11-30","2020-12-31"]
columns_to_average_2022=["2022-01-31","2022-02-28","2022-03-31","2022-04-30","2022-05-31","2022-06-30","2022-07-31","2022-08-31","2022-09-30","2022-10-31","2022-11-30","2020-12-31"]
columns_to_average_2023=["2023-01-31","2023-02-28","2023-03-31","2023-04-30","2023-05-31","2023-06-30","2023-07-31","2023-08-31","2023-09-30","2023-10-31","2023-11-30","2023-12-31"]
columns_to_average_2024=["2024-01-31","2024-02-29","2024-03-31","2024-04-30","2024-05-31","2024-06-30","2024-07-31","2024-08-31","2024-09-30","2024-10-31","2024-11-30","2024-12-31"]
```

```python
rent_df["2020"]=rent_df[columns_to_average_2020].mean(axis=1)
rent_df["2021"]=rent_df[columns_to_average_2021].mean(axis=1)
rent_df["2022"]=rent_df[columns_to_average_2022].mean(axis=1)
rent_df["2023"]=rent_df[columns_to_average_2023].mean(axis=1)
rent_df["2024"]=rent_df[columns_to_average_2024].mean(axis=1)
rent_df.head()
```

```python
mortgage_df["2020"]=mortgage_df[columns_to_average_2020].mean(axis=1)
mortgage_df["2021"]=mortgage_df[columns_to_average_2021].mean(axis=1)
mortgage_df["2022"]=mortgage_df[columns_to_average_2022].mean(axis=1)
mortgage_df["2023"]=mortgage_df[columns_to_average_2023].mean(axis=1)
mortgage_df["2024"]=mortgage_df[columns_to_average_2024].mean(axis=1)
mortgage_df.head()
```

```python
columns_to_keep=["RegionID", "RegionName", "StateName","2020", "2021", "2022", "2023", "2024"]
mortgage_clean_df = mortgage_df[columns_to_keep]
mortgage_clean_df.head()
```

```python
rent_clean_df=rent_df[columns_to_keep]
rent_clean_df.head()
```

```python
rent_clean_df.replace("",np.nan, inplace=True)
rent_clean_df.dropna(inplace=True)
rent_clean_df.reset_index().head()
```

```python
mortgage_clean_df.dropna(inplace=True)
mortgage_clean_df.replace("",np.nan, inplace=True)
mortgage_clean_df.reset_index().head()
```

```python
rent_clean_df=rent_clean_df.rename(columns= {
    "RegionID":"region_id",
    "RegionName":"region_name",
    "StateName":"state_name"})
rent_clean_df.head()
```

```python
mortgage_clean_df=mortgage_clean_df.rename(columns= {
    "RegionID":"region_id",
    "RegionName":"region_name",
    "StateName":"state_name"})
mortgage_clean_df.head()
```

```python
rent_clean_df.to_csv("output/clean_rents.csv", index=False)
mortgage_clean_df.to_csv("output/clean_mortgages.csv",index=False)
```

```python
rent_state_2020=rent_clean_df.groupby("state_name")["2020"].mean()
rent_state_2021=rent_clean_df.groupby("state_name")["2021"].mean()
rent_state_2022=rent_clean_df.groupby("state_name")["2022"].mean()
rent_state_2023=rent_clean_df.groupby("state_name")["2023"].mean()
rent_state_2024=rent_clean_df.groupby("state_name")["2024"].mean()
rent_state_df=pd.DataFrame({
    "state_name":rent_state_2020.index,
    "2020":rent_state_2020.values,
    "2021": rent_state_2021.values,
    "2022": rent_state_2022.values,
    "2023": rent_state_2023.values,
    "2024": rent_state_2024.values
})
rent_state_df.head()
```

```python
mortgage_state_2020 = mortgage_clean_df.groupby("state_name")["2020"].mean()
mortgage_state_2021 = mortgage_clean_df.groupby("state_name")["2021"].mean()
mortgage_state_2022 = mortgage_clean_df.groupby("state_name")["2022"].mean()
mortgage_state_2023 = mortgage_clean_df.groupby("state_name")["2023"].mean()
mortgage_state_2024 = mortgage_clean_df.groupby("state_name")["2024"].mean()
mortgage_state_df=pd.DataFrame({
    "state_name":mortgage_state_2020.index,
    "2020":mortgage_state_2020.values,
    "2021": mortgage_state_2021.values,
    "2022": mortgage_state_2022.values,
    "2023": mortgage_state_2023.values,
    "2024": mortgage_state_2024.values
})
mortgage_state_df.head()
```

```python
rent_state_df.to_csv("output/clean_rents.csv", index=False)
mortgage_state_df.to_csv("output/clean_mortgages.csv",index=False)
```
